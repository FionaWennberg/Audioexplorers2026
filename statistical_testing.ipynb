```python
import numpy as np
from scipy.stats import norm

def fisher_z(r):
    return 0.5 * np.log((1 + r) / (1 - r + 1e-12))

def compare_correlations(r1, r2, n):
    """
    Compare two correlations using Fisher z-transform
    """
    z1 = fisher_z(r1)
    z2 = fisher_z(r2)

    se = np.sqrt(1/(n-3) + 1/(n-3))
    z = (z1 - z2) / se

    p = 2 * (1 - norm.cdf(abs(z)))
    return z, p

def corr(a, b):
    return np.corrcoef(a, b)[0, 1]


# --- Compute correlations ---
r_front = corr(e_front, e_enh)
r_rear  = corr(e_rear, e_enh)
r_left  = corr(e_left, e_enh)
r_right = corr(e_right, e_enh)

n = len(e_enh)

print("Correlations:")
print("Front:", r_front)
print("Rear :", r_rear)
print("Left :", r_left)
print("Right:", r_right)


# --- Compare LEFT vs others ---
pairs = {
    "Left vs Front": (r_left, r_front),
    "Left vs Rear":  (r_left, r_rear),
    "Left vs Right": (r_left, r_right),
}

print("\nSignificance tests (Fisher z):")

for name, (r1, r2) in pairs.items():
    z, p = compare_correlations(r1, r2, n)
    print(f"{name}: z = {z:.2f}, p = {p:.5f}")
```
